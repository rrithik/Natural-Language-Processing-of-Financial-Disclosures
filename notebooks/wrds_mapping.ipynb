```python
import pandas as pd
import os

# Load mapping file
df = pd.read_csv("wrds_mapping.csv")

all_docs = []

print("Extracting text from WRDS filesystem...\n")

for _, row in df.iterrows():
    wrdsfname = row["wrdsfname"]

    if pd.isna(wrdsfname):
        continue

    file_path = f"/wrds/sec/warchives/{wrdsfname}"

    if not os.path.exists(file_path):
        print("MISSING:", file_path)
        continue

    try:
        with open(file_path, "r", errors="ignore") as f:
            text = f.read()

        all_docs.append(text)

        print("OK:", file_path)

    except Exception as e:
        print("FAILED:", file_path, e)


# SAVE AS ONE BIG TEXT FILE
print("\nSaving to wrds_text_data.txt...")

with open("wrds_text_data.txt", "w", errors="ignore") as out:
    for text in all_docs:
        out.write(text)
        out.write("\n\n" + "="*80 + "\n\n")

print("Done!")
```
